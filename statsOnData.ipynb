```python
import pandas
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
```

```python
dataSet = pandas.read_csv("Loan_default.csv")
print(dataSet.isnull().sum())
```

```python
print(dataSet["Default"].value_counts())
print('\n',dataSet["Default"].value_counts(normalize=True) * 100)
```

```python
print(dataSet["Education"].unique())
print(dataSet["EmploymentType"].unique())
print(dataSet["MaritalStatus"].unique())
print(dataSet["HasMortgage"].unique())
print(dataSet["HasDependents"].unique())
print(dataSet["LoanPurpose"].unique())
print(dataSet["HasCoSigner"].unique())
```

```python
# OrdinalEncoder
edu = ["High School", "Bachelor's","Master's", "PhD"]
eduEncoder = OrdinalEncoder(categories= [edu])
dataSet["Education"] = eduEncoder.fit_transform(dataSet[["Education"]])

hasMor = ["No", "Yes"]
hasMorEncoder = OrdinalEncoder(categories= [hasMor])
dataSet["HasMortgage"] = hasMorEncoder.fit_transform(dataSet[["HasMortgage"]])

hasDep = ["No", "Yes"]
hasDepEncoder = OrdinalEncoder(categories= [hasDep])
dataSet["HasDependents"] = hasMorEncoder.fit_transform(dataSet[["HasDependents"]])

hasSin = ["No", "Yes"]
hasSinEncoder = OrdinalEncoder(categories= [hasSin])
dataSet["HasCoSigner"] = hasSinEncoder.fit_transform(dataSet[["HasCoSigner"]])

#OneHotEncoding
oneHot = OneHotEncoder(handle_unknown="ignore", sparse_output= False).set_output(transform="pandas")

empEncoded= oneHot.fit_transform(dataSet[["EmploymentType"]])
dataSet = pandas.concat([dataSet, empEncoded], axis=1).drop(columns=["EmploymentType"])

marEncoded = oneHot.fit_transform(dataSet[["MaritalStatus"]])
dataSet = pandas.concat([dataSet, marEncoded], axis=1).drop(columns=["MaritalStatus"])

hasPurEncoded = oneHot.fit_transform(dataSet[["LoanPurpose"]])
dataSet = pandas.concat([dataSet, hasPurEncoded],axis=1).drop(columns=["LoanPurpose"])

print(dataSet)
```

```python
dataSet.to_csv("encoed_Loan_Default.csv", index=False)
```
