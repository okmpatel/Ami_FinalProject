```python
import pandas
from sklearn.preprocessing import OrdinalEncoder
```

```python
dataSet = pandas.read_csv("Loan_default.csv")
print(dataSet.isnull().sum())
```

```python
print(dataSet["Default"].value_counts())
print('\n',dataSet["Default"].value_counts(normalize=True) * 100)
```

```python
print(dataSet["Education"].unique())
print(dataSet["EmploymentType"].unique())
print(dataSet["MaritalStatus"].unique())
```

```python
edu = ["High School", "Bachelor's","Master's", "PhD"]
eduEncoder = OrdinalEncoder(categories= [edu])
dataSet["Education"] = eduEncoder.fit_transform(dataSet[["Education"]])

emp = ['Unemployed', 'Self-employed', 'Part-time','Full-time']
empEncoder = OrdinalEncoder(categories= [emp])
dataSet["EmploymentType"] = empEncoder.fit_transform(dataSet[["EmploymentType"]])

mar = ['Single', 'Married', 'Divorced']
marEncoder = OrdinalEncoder(categories= [mar])
dataSet["MaritalStatus"] = marEncoder.fit_transform(dataSet[["MaritalStatus"]])

dataSet.head(100)
```
